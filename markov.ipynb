```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


rho = 0.5
N =100
Ensemble = []
L = 10
t = []
i = []
for ii in range(L):
    sites = np.ones(N)
    sites[:int(rho*N)] = 0
    np.random.shuffle(sites)
    Ensemble.append(sites)
    i.append(ii)
    #plt.imshow(pd.DataFrame(Ensemble))  
#print(Ensemble[1])
#Ensemble = np.array(Ensemble)
plt.imshow(Ensemble)
Ensemble[1][1]
```

```python
ensemble = []
for b in range(40):
    for o in i:
        Ensemble[o][99] = Ensemble[o][1]
        for q in range(99):
            if Ensemble[o][q ] == 1 :
                if Ensemble[o][q + 1] == 0:
                    Ensemble[o][q + 1] = 1
                    Ensemble[o][q] = 0
        ensemble.append(Ensemble)

    plt.imshow(ensemble[1])
    plt.show()
```
